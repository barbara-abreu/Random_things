# Analysis of the proportions of conformations in PCA with other components

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load dataframes
combs=[(i, k) for i in [1,2] for k in range(3,11)]
d={}
for i in combs:
    namedf='PC_'+str(i[0])+'_'+str(i[1])
    namep='proportions_2d_pc_'+str(i[0])+'_'+str(i[1])+'_0.02.csv'
    d[namedf]=pd.read_csv(namep, delimiter=";")
```

```python
#Drop unused columns
names_df=list(d.keys())
for name, df in d.items():
    d[name]=d[name].drop(['#ATP', '#ADP', '#Apo'],axis=1)
```

```python
#Plot stuff
tmp={}
for name, df in d.items():
    tmp[name]=d[name].drop('Basin',axis=1)

```

```python
def plot_barchart(impdata, tit):
    fig=plt.figure(figsize=(6,3))
    ax=fig.add_axes([0,0,1,1])
    x=list(impdata.index)
    width=0.3
    xx=list(map(lambda x:x-0.3, x))
    xxx=list(map(lambda x:x+0.3, x))

    ax.bar(xx, impdata['%ATP'], width, label='ATP', color='crimson')
    ax.bar(x, impdata['%ADP'], width, label='ADP', color='royalblue')
    ax.bar(xxx, impdata['%Apo'], width, label='Apo', color='limegreen')
    ax.set_xlabel("#Basin")
    ax.set_ylabel("Percentage")
    ax.set_title(tit)
    ax.legend()

```

```python
for name, df in tmp.items():
    plot_barchart(tmp[name], name)

```
